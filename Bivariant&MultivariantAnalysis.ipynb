```python

#numerical to numerical bivariant
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])

```

```python
#multivariant
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['smoker'])
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python

import matplotlib.pyplot as plt

sns.scatterplot(
    data=tips,             # dataframe
    x='total_bill',        # x-axis column name
    y='tip',               # y-axis column name
    hue='sex',             # color by sex
    style='smoker',        # marker by smoker
    size='size',           # size of points
    palette='Set1',        # optional: nice colors
    sizes=(20,200)         # optional: range of point sizes
)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='pclass', y='age', data=titanic)
plt.show()
```

```python
sns.barplot(x='pclass', y='fare', data=titanic)
```

```python
sns.barplot(
    x='pclass',
    y='age',
    hue='sex',
    data=titanic
)
plt.show()
```

```python
#numercal-to-categorical BOx Plot

sns.boxplot(
    x='sex',
    y='age',
    hue='survived',
    data=titanic
)
plt.show()


```

```python


sns.kdeplot(
    data=titanic[titanic['survived'] == 0],
    x='age'
)
sns.kdeplot(
    data=titanic[titanic['survived'] == 1],
    x='age'
)
plt.show()
```

```python
#categorical & Categorical Analysis
titanic.head()
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
titanic.groupby('embarked')['survived'].mean() * 100

```

```python
#ClusterMap (Categorical - Categorical)
sns.clustermap(pd.crosstab(titanic['pclass'],titanic['survived']))
```
