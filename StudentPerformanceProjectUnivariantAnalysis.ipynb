```python
df.columns
```

```python
df['math score'].describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['math score'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.boxplot(df['math score'])
plt.title('Boxplot of Math Score')
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(df['math score'], fill=True, color='green')
plt.title('Density of Math Score')
plt.show()
```

```python
print("Mean:", df['math score'].mean())
print("Median:", df['math score'].median())
print("Mode:", df['math score'].mode()[0])
print("Standard Deviation:", df['math score'].std())
print("Min:", df['math score'].min())
print("Max:", df['math score'].max())
```

```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['math score'] < Q1 - 1.5*IQR) | (df['math score'] > Q3 + 1.5*IQR)]
print("Outliers:\n", outliers)
```
