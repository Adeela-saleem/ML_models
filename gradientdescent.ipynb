```python
import numpy as np  #for numbers & math 
import matplotlib.pyplot as plt  #for graph
```

```python
np.random.seed(42)
```

```python
#area
X=np.random.rand(100,1) #1 col ,100 rows

#prcie 

y = 4 + 3 * X + np.random.randn(100, 1)  #-->y=mx+c
print(X.shape)
print(y.shape)

```

```python

print("Pehle 5 areas:", X[:5].flatten().round(2))
print("Pehle 5 prices:", y[:5].flatten().round(2))
```

plotting data

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.5)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.grid(True)
plt.show()
```

```python

theta = np.random.randn(2, 1)  # 2 weights: bias + slope
X_b = np.c_[np.ones((100, 1)), X]  # bias column add karo

print("Starting weights:", theta.flatten().round(3))
print("X_b shape:", X_b.shape)
print("X_b 1st three rows")
print(X_b[:3])
```

```python
# Hyperparameters
learning_rate = 0.1
n_iterations = 1000
n = 100  # samples

# Loss store 
loss_history = []

# Gradient Descent Loop
for iteration in range(n_iterations):
    
    # Step 1: Predict 
    y_pred = X_b.dot(theta)
    
    # Step 2: Error n
    error = y_pred - y
    
    # Step 3: Gradient nikala
    gradient = X_b.T.dot(error) / n
    
    # Step 4: Weights update 
    theta = theta - learning_rate * gradient
    
    # Loss save 
    loss = np.mean(error**2)
    loss_history.append(loss)

print("Final weights:", theta.flatten().round(3))
print("Actual weights: bias=4, slope=3")
print("Final Loss:", loss_history[-1].round(4))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(loss_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss Har Iteration Mein Kam Hota Gaya!')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# Actual data
plt.scatter(X, y, color='blue', 
            alpha=0.5, label='Actual')

# GD ki predicted line
X_line = np.linspace(0, 2, 100).reshape(-1, 1)
X_line_b = np.c_[np.ones((100, 1)), X_line]
y_line = X_line_b.dot(theta)

plt.plot(X_line, y_line, color='red', 
         linewidth=2, label='GD Prediction')

plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Gradient Descent — Best Fit Line!')
plt.legend()
plt.grid(True)
plt.show()
```
